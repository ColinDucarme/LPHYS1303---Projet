```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
%matplotlib widget
```

```python
Jup_true = np.loadtxt(fname=".\Data\Jupiter.txt",dtype=float,delimiter=',',skiprows=5,usecols=(2,3,4))
Sat_true = np.loadtxt(fname=".\Data\Saturne.txt",dtype=float,delimiter=',',skiprows=5,usecols=(2,3,4))

Jup_calc = {}
Sat_calc = {}
Jup_calc['Heun'] = np.loadtxt(fname="./Data/Jup_pos_Heun.txt",dtype=float,delimiter=',')
Jup_calc['SV'] = np.loadtxt(fname="./Data/Jup_pos_SV.txt",dtype=float,delimiter=',')
Sat_calc['Heun'] = np.loadtxt(fname="./Data/Sat_pos_Heun.txt",dtype=float,delimiter=',')
Sat_calc['SV'] = np.loadtxt(fname="./Data/Sat_pos_SV.txt",dtype=float,delimiter=',')
```

#### Heun

```python
fig= plt.figure(figsize=(10,7))
ax1 = fig.add_subplot(1,1,1,projection='3d')
ax1.plot(Jup_calc['Heun'][:,0],Jup_calc['Heun'][:,1],Jup_calc['Heun'][:,2],c='blue',linewidth=.05)
ax1.plot(Jup_true[:,0],Jup_true[:,1],Jup_true[:,2],c='green',linewidth=.05)

plt.show()
```

#### SV

```python
fig= plt.figure(figsize=(10,7))
ax1 = fig.add_subplot(1,1,1,projection='3d')
ax1.plot(Jup_calc['SV'][:,0],Jup_calc['SV'][:,1],Jup_calc['SV'][:,2],c='blue',linewidth=.05)
ax1.plot(Jup_true[:,0],Jup_true[:,1],Jup_true[:,2],c='green',linewidth=.05)

plt.show()
```
